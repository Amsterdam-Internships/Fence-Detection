```python
import os
import cv2

import numpy as np

from matplotlib import pyplot as plt

img = cv2.imread(os.path.join('..', 'data', 'pano_0000_000015_l.jpg'), 0)
edges = cv2.Canny(img,300,0)

plt.rcParams["figure.figsize"] = (20,10)
plt.imshow(edges,cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.show()
```
