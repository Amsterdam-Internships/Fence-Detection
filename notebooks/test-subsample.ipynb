```python
import os
import cv2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n = 1000
root = os.path.join('..', 'data', 'images')
```

```python
np.random.seed(2)
# np.random.seed(1) # used for spectrum intelligence sample

metadata = pd.read_csv(os.path.join(root, 'metadata.csv'))
indices = np.random.choice(range(0, metadata.shape[0]), n, replace=False)
print(len(indices))
```

```python
subsample = metadata.iloc[indices]
print(subsample.shape)
subsample.to_csv(os.path.join(root, 'sample-metadata.csv'), index=False)
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.scatter(metadata.lng, metadata.lat)
plt.scatter(subsample.lng, subsample.lat, marker='x', color='r')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (10,5)
for _, row in subsample.iterrows():
    img = cv2.imread(os.path.join(root, 'images', f'{row["filename"]}.jpg'))
#     plt.imshow(img)
#     plt.show()
    cv2.imwrite(os.path.join(root, 'sample-1000', f'{row["filename"]}.jpg'), img)
```
